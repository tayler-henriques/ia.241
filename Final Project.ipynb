### Covid-19 is a contagious disease that causes respiratory distress as well as numerous other symptoms. The first known case was discovered in Wuhan China in December of 2019 and since then has become a global pandemic. This virus affects all ages and has caused many countries to enter a shut down. The purpose of the shut down was to stop the spread of Covid-19 while researchers were continuting to learn more about the virus and attempting to create a vaccine. In Portugal the virus did not hit many of its citizens early on, instead they are beginning to feel the full effects until January of 2021 when their cases rose tremendously. As of May 2021 they are able to keep their cases to a low number and offer a vaccine to its citizens. 

```python
%matplotlib inline
import pandas
```

```python
df=pandas.read_excel('s3://tayler-ia241/COVID-19-geographic-disbtribution-worldwide-2020-12-14-2.xls')
df[:5]
```

## Question 1: How many confirmed Covid cases does Portugal have?

```python
portugal_data=df.loc[df['countriesAndTerritories']=='Portugal']
portugal_data[:10]
```

```python
portugal_cases=portugal_data.groupby('dateRep').sum()['cases']
portugal_cases.plot()
```

## Question 2: How many covid related deaths are there in Portugal?

```python
portugal_deaths=portugal_data.groupby('dateRep').sum()['deaths']
portugal_deaths.plot()
```

## Question 3: How do the cases in Portugal compare to the cases in the rest of Europe?

```python
europe_data=df.loc[df['continentExp']=='Europe']
europe_data[:10]
```

```python
europe_cases=europe_data.groupby('countriesAndTerritories').sum()['cases']
europe_cases.nlargest(15).plot.bar()
```

### Portugal has the 15th highest number of cases out of the 50 countries in Europe.

#### Conclusion: In conclusion, Portugal has had a large number of Covid cases as well as a very high number of deaths due to covid especially in the recent months. They do rank 15th for the highest number of cases in Europe.
